```python
# Q1

def introduce(name,age):
    return f"My name is {name} and I am {age} years old."
print(introduce('Rhim',28))
```

```python
#Q2
def batch_generator(my_list,batch_size):
    for i in range(0,len(my_list),batch_size):
        yield my_list[i:i+batch_size]
x=batch_generator([1,2,3,4,5,6,7,8],3)
print(next(x))
print(next(x))
print(next(x))
```

```python
def square_list(nums):
    return list(map(lambda x: x**2, nums))

numbers = [1, 2, 3, 4]
print(square_list(numbers))
```

```python
#Q3
import pandas as pd

numbers=[1,2,3,4]
df=pd.DataFrame(numbers,columns=['num'])

df['square']=df['num'].map(lambda x:x**2)
print(df['square'])
lst=df['square'].tolist()
print(lst)
```
