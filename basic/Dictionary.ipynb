# Dictionary 

```python
# key: value pair
#duplicate value not allowed 
#muteable 
#unorder 
info={
    "name":"Asad",
    "Cgpa":3.30,
    "Subjects":["math","science"]
}
print(type(info))
```

```python
info={
    "name":"Asad",
    "Cgpa":3.30,
    "Subjects":["math","science"]
}
print(info['name'])
```

```python
info={
    "name":"Asad",
    "Cgpa":3.30,
    "Subjects":["math","science"]
}
print(info[])
```

# Dictionary method 

```python
#keys
info={
    "name":"Asad",
    "Cgpa":3.30,
    "Subjects":["math","science"]
}
print(info.keys())
print(info.values())
```

```python
info={
    "name":"Asad",
    "Cgpa":3.30,
    "Subjects":["math","science"]
}
dict_keys=list(info.keys())
print(dict_keys)
print(type(dict_keys))
```

```python
#items
info={
    "name":"Asad",
    "Cgpa":3.30,
    "Subjects":["math","science"]
}

print(info.items())
```

```python
#get 
# if item not present in dict then show none value 
info={
    "name":"Asad",
    "Cgpa":3.30,
    "Subjects":["math","science"]
}
print(info.get("name"))
print(info.get("name2"))
```

```python
#update method 
info={
    "name":"Asad",
    "Cgpa":3.30,
    "Subjects":["math","science"]
}
info.update({
    "city":"Dhaka"
})
print(info)
```
