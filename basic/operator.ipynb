### operator 

```python
a=9
b=5
print(a+b)
print(a-b)
print(a*b)
print(a/b)
print(a%b) #modulo
```

### Relational Operator 

```python
a=10
b=5
print(a<b)
print(a>b)
print(a==b)
print(a !=b)
```

### Assignment operator 

```python
a=10
a=a+5 #a=15
a-=5 #a=a-5
a%=5
print(a)
```

### Logical operator 


```python
var= False 
print(not var) # not operator 

print(not(5>8))
```

```python
print((5>3)and (3>8)) # and operator 
print((5>3)and (3<8))
print((5>3)or (3<8)) # operator 
```
