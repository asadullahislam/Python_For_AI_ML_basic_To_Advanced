# Tuple

```python
#tuple is immutable in pyton and value not assignable 
tup=(1,2,3,4,5)
print(tup)

print(type(tuple))
```

```python
# single value is not tuple 
tup=(1)
print(type(tup))
```

```python
tup=(1,)
print(type(tup))
```

```python
tup=(1,2,3,4,5)
print(tup[0:3])
```

# Tuple method 

```python
tup=(1,2,3,4,5)
for val in tup:
    print(val)
```

```python
tup=(1,2,3,4,5)
sum=0
for val in tup:
    sum+=val
print(f"sum of val is {sum}")
```

```python
#index method 

tup=(1,2,2,3,2,4,5)

print(tup.index(2))
```

```python
#count method
tup=(1,2,3,4,2,2,5)
print(tup.count(2))
```
