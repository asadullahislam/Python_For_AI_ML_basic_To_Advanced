### Conditional Statements Match case , this is alternate for if, elif, else case

```python
color = input("enter color : ")

match color:
    case "Green":
        print("Go")
    case "yellow":
        print("Look")
    case "red":
        print("stop")
    case _:
        print("wrong color")
```
