```python
# print factorial of "n"
def calc_factorial(n):
    fact=1
    for i in range(1,n+1):
        fact*=i
    return fact
n= int(input("Enter n: "))
print(calc_factorial(n))

```

# Different type of  Arguments


```python
# Positional Arguments 
def adder(a,b,c):
    print(f"the nums are : {a},{b},{c}")
adder(5,10,20)
```

```python
#variable size Arguments

def adder(*args):
    print(args)
adder(2,3,4,5,6,7)
```

```python
#variable size Arguments

def adder(*args):
    for i in args:
        print(i)
adder(2,3,4,5,6,7)
```

# Keyword Arguments

```python
#keyword arguments


def adder(a,b,c):
    print(f"{a},{b},{c}")
adder(a=3,b=6,c=9)
```

```python
#optional argument


def adder(a=3,b=6,c=9):
    print(f"{a},{b},{c}")
adder(a=3,b=6,c=10)
```

```python

def adder(a=3,b=6,c=9):
    print(f"{a},{b},{c}")
adder()
```

```python

def adder(**args):
    print(args)
adder(a=3,b=6,c=10)
```

```python

def adder(**args):
    for key,value in args.items():
        print(f"{key}:{value}")
    
adder(a=3,b=6,c=10)
```

# Lambda , Map  and Apply Function 

```python
## Apply and lambda function 

import pandas as pd
df=pd.DataFrame({'temperature (F)':[32,212,98.6,68]})
df['temperature (C)']=df['temperature (F)'].apply(lambda x:(x-32)*5/9)
print(df)
```

# Map & lambda function 

```python
import pandas as pd

df=pd.DataFrame({'A':[1,2,3,4],'B':[5,6,7,8]})
print(df)
df['A']=df['A'].map(lambda x:x**0.5)
print(df)
```

```python
#apply and lambda

import pandas as pd 
df=pd.DataFrame({'total_spent':[120,600,450,800]})

# Creating a 'High value' featur on the fly 
df['is_high_value']=df['total_spent'].apply(lambda x: 1 if x>500 else 0)
print(df)
```
