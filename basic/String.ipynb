# slice string


```python
# str[start idx: end idx]
word="Python"
print(word[2:4])
```

```python
word="I study in Universiy "
print(word[5:13])
print(word[11:])
print(word[11:len(word)])
```

```python
print(word[:])
print(word[:13])
```

# Strings formatting

```python
#format(),f-string
```

```python
a=5
b=10
sum=a+b
# normal formating 
print("Language is ".format("Python"))
print("sum is {}".format(sum))
```

```python
print("sum of {} & {} is {}".format(a,b,sum))
```

```python
#index based formatting
a=5
b=10
sum= a+b
print("sum of {1} & {0} is {2}".format(a,b,sum))
```

```python
# value based formatting
print("values of vars {a} & {b}".format(a=3,b=10))
```

```python
#f-string , Liter String interpolation 
a=5
b=10
print(f"Sum of {a} & {b} is {a+b} and avg is {(a+b)/2}")
```
