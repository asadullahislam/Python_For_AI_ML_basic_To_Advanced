# Slicing Lists


```python
# str[st idx: end idx]
# st idx: default =0 , end idx = default is len(str)
marks=[99,89,100,65,"abc",92]
print(marks[5:len(marks)])
```

```python
#negativ index
marks=[99,89,100,65,92,"abc",100.99]
print(marks[-5:-2])
```

# Lists methods

```python
# I.append(v) # add one elemetn at the end 
# I.insert(idx,val) insert element at idx
#I.sort() arranges in increasing order 
#l.reverse() # reverse order
```

```python
#append
numbers=[1,2,3]
numbers.append(4)
print(numbers)
```

```python
#insert
numbers=[1,2,3]
numbers.insert(2,10)
print(numbers)
```

```python
# sort

numbers=[3,4,5,2,1]
numbers.sort()
print(numbers)
numbers.sort(reverse=True)
print(numbers)
```

```python
#reverse
numbers=[3,4,5,2,1]
numbers.reverse()
print(numbers)
```

# Loops in list 

```python
nums=[1,2,3,10,4]
x=10
idx=0
for val in nums:
   if(val==x):
        print(f"{x} found at index={idx}")
        break
   idx+=1
```
