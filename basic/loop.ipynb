## Break & continue

```python
i=1
while(i <=10):
    if i%6==0:
        break
    print(i)
    i+=1 # update on 
print("outside loop now...")
```

```python

# continue skip current iteration and go next iteration 
i=1
while(i <=10):
    if (i % 3==0):
        i+=1
        continue
    print(i)
    i+=1
print("Outside loop now..")
```

```python
#odd nums
i=1
while (i<=10):
    print(i)
    i+=2
```

```python
#odd nums
i=1
while (i<=10):
    if (i % 2==0):
        i+=1
        continue
    print(i)
    i+=1
```

# For loops

```python
# for use sequential traversal like lsit, tuple, dictionary, string , sequence , sequence of character 

string ="Hello"
#in => membership operator (to check presence)
for val in string: 
    print(val,end=" ")

   
```

```python
# sequence 
#range(n)->>>>0 to (n-1)
for i in range(10):
    print(i)
```

```python
for i in range(10):
    print(i+1)
```

```python
for i in range (5):
    print("hello world")
```

```python
word ="Artificial inteligence"
# count the number of i's

count=0
for i in word:
    if i=='i':
        count=count+1
print(count)
```

```python
#vowel count 
word ="Artificial inteligence"
count=0

for i in word:
    if (i=='a' or i=='e' or i=='i' or i=='o' or i=='u'):
        count+=1
print(count)
```

## loop range() function 

```python
#range(5) means 0 to 4
#range(n) means 0 to (n-1)
#range(start,stop,step), default start=0, step =1 by default behaviour
#range(1,6)--> 1,2,3,4,5
#range(1,10,2) ->>1,3,5,7,9

for i in range(5,10):
    print(i)
```

```python
for i in range(1,10,2):
    print(i)
```

```python
for i in range(2,11,2):
    print(i)
```

```python
#print sum of first'n' natural numbers
n= int(input('Enter number'))
sum =0
for i in range(1,n+1):
    sum+=i
print(sum)
```
