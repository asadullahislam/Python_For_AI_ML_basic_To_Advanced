```python
def even_num(a, b):
    print("Numbers are:", a, b)

    # Check both numbers separately
    if a % 2 == 0:
        print(f"{a} is Even")
    else:
        print(f"{a} is Odd")

    if b % 2 == 0:
        print(f"{b} is Even")
    else:
        print(f"{b} is Odd")

even_num(5,2)
```

```python
def print_digits(n):
    if n == 0:
        return
    print_digits(n // 10)
    print(n % 10, end=" ")
print_digits(321)
```

```python
def print_digits(n):
    digits = []

    while n > 0:
        digits.append(n % 10)
        n //= 10

    # reverse to correct order
    for d in reversed(digits):
        print(d, end=" ")
print_digits(321)
```

```python
def print_digits(n):
    if n > 0:
        print_digits(n // 10)  # Go deeper until n is 0
        print(n % 10, end=" ") # Print on the way back out

print_digits(312)  # Output: 3 1 2
```

```python
def print_digit(n):
    if n>0:
        print_digit(n//10)
        print(n%10 ,end=" ")
print_digit(321) 
```
